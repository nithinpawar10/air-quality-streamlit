```python
df.drop(columns=['Date', 'Time'], inplace=True)
```

```python
def classify_pollution_level(co):
    if co <= 1:
        return "Good"
    elif co <= 2.5:
        return "Moderate"
    elif co <= 5:
        return "Unhealthy"
    else:
        return "Hazardous"

df['PollutionLevel'] = df['CO(GT)'].apply(classify_pollution_level)
df['PollutionLevel'].value_counts()
```
